```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('sample_data/laxmi.jpeg', 0)
cv2_imshow(img)
img = cv2.resize(img, (300, 300))
cv2_imshow(img)
img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(img)
cv2.imwrite('sample_data/laxmi_new.jpeg', img)
```
